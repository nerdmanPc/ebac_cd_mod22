```python
ativos = ['U', 'NVDA', 'TSLA']
dt_ini = '01/01/2012'
```

```python
yf = web.get_data_yahoo(ativos, start=dt_ini)
yf.to_csv('yahoo_finance.csv')
```

#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
amp_columns = pd.MultiIndex.from_tuples(tuples=[
    ('Amplitude', 'U'),
    ('Amplitude', 'NVDA'),
    ('Amplitude', 'TSLA')
])
amplitude = (yf['High'] - yf['Low'])
amplitude.columns = amp_columns
yf = yf.join(amplitude)
```

```python
amp_vertical = yf['Amplitude'].stack().reset_index()
amp_vertical.columns = ['Data', 'Ativo', 'Amplitude']
amp_vertical
```

```python
fig = px.line(
    data_frame=amp_vertical,
    x='Data',
    y='Amplitude',
    color='Ativo',
    template='plotly_white'
)
fig.show()
```

A Tesla parece ter tido a maior volatilidade absoluta nos últimos meses.
